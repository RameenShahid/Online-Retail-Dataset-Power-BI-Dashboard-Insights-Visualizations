# Feature Engineering for Customer Sales & Behavior Analysis ðŸš€

##### Recency, Purchase Frequency, CLV â†’ Behavioral Features (Understand customer habits)
##### Monetary Value, AOV, Revenue Per Customer â†’ Financial Features (Track spending patterns)
##### Day of the Week, Discounted Orders, Avg Items Per Order â†’ Transactional Features (Analyze shopping trends)
##### Customer Segmentation (VIP, High, Low, etc.) â†’ Categorical Features (Useful for personalization & machine learning)

```python
import pandas as pd
import numpy as np
# Load Data
df = pd.read_csv("Cleaned_Online_Retail.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df.isnull().sum()
```

```python
df['CustomerID'] = df['CustomerID'].astype(int)
```

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

```python
df.head()
```

```python
df['Month'] = df['InvoiceDate'].dt.strftime('%B')

```

```python
df.head()
```

## Monetary Analysis
##### Monetary Value (Total Spend per Customer) â€“ Total amount spent by each customer.
##### Revenue Per Customer â€“ Total revenue generated by each customer.
##### Average Order Value (AOV) â€“ The average amount spent per order.

```python
df['MonetaryValue'] = df.groupby('CustomerID')['TotalPrice'].transform('sum')
```

```python
df.head()
```

```python
df['Target'] = (df['MonetaryValue'] > df['MonetaryValue'].median()).astype(int)
```

```python
df.head()
```

```python
df['PurchaseFrequency'] = df.groupby('CustomerID')['InvoiceNo'].transform('nunique')
```

```python
df.head()
```

```python
df['AverageOrderValue'] = df['MonetaryValue'] / df['PurchaseFrequency']
```

```python
df.head()
```

```python
df['CLV'] = df['AverageOrderValue'] * df['PurchaseFrequency']
```

```python
df.head()
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])  # Ensure datetime format
df['DayOfWeek'] = df['InvoiceDate'].dt.day_name()  # Get full day name (e.g., 'Monday')
```

```python
df.head()
```

```python
df['AvgItemsPerOrder'] = df.groupby('InvoiceNo')['Quantity'].transform('sum') / df.groupby('InvoiceNo')['InvoiceNo'].transform('count')
```

```python
df['DiscountedOrders'] = (df['UnitPrice'] < 0).astype(int)
```

```python
df.head()
```

```python
df['CustomerSegment'] = pd.qcut(df['MonetaryValue'], q=4, labels=['Low', 'Mid', 'High', 'VIP'])
```

```python
df.head()
```

```python
df.to_csv("processed_Online_Retail.csv", index=False)
```
